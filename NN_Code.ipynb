The cheese case

```python
# Let's start the example discussed in lecture

data = {'Obs.': [1, 2, 3, 4, 5, 6],
    'Fat': [0.2, 0.1, 0.2, 0.2, 0.4, 0.3],
    'Salt': [0.9, 0.1, 0.4, 0.5, 0.5, 0.8],
    'Acceptance': ['like', 'dislike', 'dislike', 'dislike', 'like', 'like']}

df = pd.DataFrame(data)
df
```

```python
# define the feature set and the target set

X = df[['Fat', 'Salt']]
y = df['Acceptance']
```

```python
classes = sorted(y.unique())
classes
```

```python
# construct a neural network with a single hidden layer with 3 nodes

clf = MLPClassifier(hidden_layer_sizes=(3), activation='logistic', solver='lbfgs', random_state=1)

# fit the model to the data and predict the target values
clf.fit(X, y)
clf.predict(X)
```

```python
# Network structure and estimated weights
print('Intercepts')
print(clf.intercepts_)
print('Weights')
print(clf.coefs_)

# Prediction
print(pd.concat([df,pd.DataFrame(clf.predict_proba(X), columns=classes)], axis=1))
```

```python
epochs = clf.n_iter_
print("Number of epochs:", epochs)
```

```python
url = "https://raw.githubusercontent.com/yiding2022/DAAI/main/accidentsnn.txt"
accidents_df = pd.read_csv(url)
```

```python
accidents_df
```

```python
# read data

# accidents_df = pd.read_csv('accidentsnn.csv')
accidents_df.SUR_COND = accidents_df.SUR_COND.astype('category')
accidents_df.MAX_SEV_IR = accidents_df.MAX_SEV_IR.astype('category')
```

```python
accidents_df
# ALCHL_I Presence (1) or absence (2) of alcohol
# PROFIL_I_R Profile of the roadway: level (1), other (0)
# SUR_COND Surface condition of the road: dry (1), wet (2), snow/slush (3), ice (4), unknown (9)
# VEH_INVL Number of vehicles involved
# MAX_SEV_IR Presence of injuries/fatalities: no injuries (0), injury (1), fatality (2) - outcome variable
```

```python
# convert the categorical data into dummy variables
# exclude the column for SUR_COND 9 = unknown
processed = pd.get_dummies(accidents_df, columns=['SUR_COND'])
processed = processed.drop(columns=['SUR_COND_9'])
outcome = 'MAX_SEV_IR'
predictors = [c for c in processed.columns if c != outcome]
```

```python
processed
```

```python
predictors
```

```python
# partition data
X = processed[predictors]
y = processed[outcome]
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
# train neural network with 2 hidden nodes
clf = MLPClassifier(hidden_layer_sizes=(7), activation='logistic', solver='lbfgs', random_state=1)
clf.fit(train_X, train_y.values)
```

```python
# training performance (use idxmax to revert the one-hot-encoding)
classificationSummary(train_y, clf.predict(train_X))
# validation performance
classificationSummary(valid_y, clf.predict(valid_X))

# We see that classes 0 and 1
# are perfectly classified in both the training and validation data. However, class 2 has a high rate of misclassification.
```

```python
print('Intercepts')
print(clf.intercepts_)
print('Weights')
print(clf.coefs_)
print('# of Epochs:', clf.n_iter_)
```

Avoid overfitting

```python
# let's limit the number of epochs to 100

clf = MLPClassifier(hidden_layer_sizes=(7), activation='logistic', solver='lbfgs', random_state=1, max_iter=100)
clf.fit(train_X, train_y.values)
```

```python
# training performance (use idxmax to revert the one-hot-encoding)
classificationSummary(train_y, clf.predict(train_X))
# validation performance
classificationSummary(valid_y, clf.predict(valid_X))

# performance on validation data improved, but the performance on the training data is similar.
```
